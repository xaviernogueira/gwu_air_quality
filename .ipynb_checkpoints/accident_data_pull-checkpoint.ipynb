### Bring in data, remove extra columns, keep only 2019 accidents

```python
# bring in data
out_folder = r'C:\Users\xrnogueira\Documents\Data\Traffic_accidents'
in_df = pd.read_csv(out_folder + '\\raw_US_Accidents_Dec20.csv')
in_df.head()
```

```python
# remove extra columns
print('Column headers: %s' % in_df.columns)
keep_cols = ['Start_Time', 'Start_Lat', 'Start_Lng',  'State']
in_df = in_df[keep_cols]
in_df.head()
```

```python
# see all states and delete data points outside of the continental USA
print('States: %s' % in_df.State.unique())
```

```python
# make a dataframe with just pertinent datat and create new month, day of the week, weekend/weekday columns
time_col = 'Start_Time'

in_df[time_col] =  pd.to_datetime(in_df[time_col])
in_df['year'] = in_df[time_col].dt.year
in_df['month'] = in_df[time_col].dt.month
in_df['dayofmonth'] = in_df[time_col].dt.day
in_df['dayofyear'] = in_df[time_col].dt.dayofyear
in_df['hour'] = in_df[time_col].dt.hour

# create column coding days as either weekdays or weekends (1=weekend, 0=not weekend)
in_df['dayname'] = in_df[time_col].dt.day_name()
in_df['weekend'] = np.where((in_df['dayname'] == 'Saturday') | (in_df['dayname'] == 'Sunday') , 1, 0)
in_df.drop(['dayname'], axis=1, inplace=True)

# drop Alaska, Hawaii, and Puerto Rico (| is or in pandas)
in_df.drop(in_df[(in_df.State == 'AK') | (in_df.State == 'HI')].index, inplace=True)
in_df.head()
```

```python
# make new dataframe with only 2019 data and proper formatting
data = in_df.rename({'Start_Time':'time', 'Start_Lat':'lat', 'Start_Lng':'long'}, axis=1)
data = data.loc[data['year'] == 2019]
data19 = data[['time', 'lat', 'long']]
data19.head()
```

```python
# save to all accidents and 2019 accidents to clean csv files
data.to_csv(out_folder + '\\all_accidents.csv')
data19.to_csv(out_folder + '\\2019_accidents.csv')
```
